```python
def Distribution(Gdis):
   
    n = []
    mj = np.unique(Gdis)
    for ii in mj:
        n.append(len(np.where(Gdis>=(ii))[0]))
    y = [(n[t]/n[0]) for t in range(len(n))]
        
    return mj,y,n
```

```python
a0,a1,a2,a3,a4=[],[],[],[],[]
h= 0
with open('communities_degrees_Feb23-29.csv', newline = '') as players_data: #communities_degrees_Feb23-29 Apr05-11
    game_reader = csv.reader(players_data, delimiter=',')
    for game in game_reader:
        if h==0: print(game)
        if h>1:
            a0.append(game[0])
            a1.append(int(game[1]))
            a2.append(int(game[2]))
            a3.append(float(game[3]))
            a4.append(float(game[4]))
        h += 1
```

```python
df = pd.DataFrame({'a':a1, 'b':a2})
m_s = df.groupby('a')['b'].apply(list)
```

```python
r= 0
fig = pl.figure(figsize=(5.3,5))
for i in m_s:
    xk,yk,nk = Distribution(np.array(i))
    pl.plot(xk, nk,'*',color=color_feb[r],linewidth= 3,markersize=2)
    r+=1
pl.xlabel('Degree',fontsize=20)
pl.ylabel('Nodes',fontsize=20)
pl.xscale('log')
pl.yscale('log')
pl.xticks(fontsize=16)
pl.yticks(fontsize=16)
pl.tight_layout()
```
